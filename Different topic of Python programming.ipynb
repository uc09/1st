In Python's **string formatting** using the ``%`` operator, the first `%` is used to introduce a format specifier, while the second `%` acts as the operator that connects the string and the value(s) to be formatted. Let me explain how this works in your example:

    print("Mean absolute error: %.2f" % mean_value)

Here’s the breakdown:

- "`Mean absolute error: %.2f`":
    This is the format string, where `%.2f` is the format specifier. It tells Python how to format the value (in this case, it will format the value as a floating-point number with two decimal places).

- `% mean_value`:
    This is where the value (mean_value) is provided. The `%` operator is used to **interpolate** (insert) `mean_value` into the format string at the location of `%.2f`.
    
Why Two %?

- The **first %** is inside the string, indicating that a value will be inserted in place of `%.2f`, and also tells Python how to format that value.
- The second `%` outside the string is the operator that links the format string to the value you're formatting.



In summary, the **first** `%` is part of the string to indicate where and how a value will be inserted, and the **second** `%` outside the string is the operator that links the string to the actual value(s) being formatted.

##### Example : BankAccount Class

Here's a more complex example with a BankAccount class where methods interact with each other using `self`.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  # Private attribute
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}.")
        else:
            print("Deposit amount must be positive.")
        # Call to get_balance to display updated balance
        self.display_balance()
    
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}.")
        else:
            print("Insufficient funds.")
        # Call to get_balance to display updated balance
        self.display_balance()
    
    def get_balance(self):
        return self.__balance
    
    def display_balance(self):
        # Calling get_balance method to retrieve the current balance
        balance = self.get_balance()
        print(f"Current balance: {balance}")
    
    def transfer(self, amount, target_account):
        if amount <= self.__balance:
            self.withdraw(amount)  # Call withdraw method
            target_account.deposit(amount)  # Call deposit method on another instance
            print(f"Transferred {amount} to {target_account.owner}.")
        else:
            print("Transfer failed: Insufficient funds.")
        # Optionally, display both balances
        self.display_balance()
        target_account.display_balance()

# Create two BankAccount instances
account1 = BankAccount("Alice", 1000)
account2 = BankAccount("Bob", 500)

# Perform a deposit
account1.deposit(200)
# Output:
# Deposited 200.
# Current balance: 1200

# Perform a withdrawal
account2.withdraw(100)
# Output:
# Withdrew 100.
# Current balance: 400

# Perform a transfer from Alice to Bob
account1.transfer(300, account2)
# Output:
# Withdrew 300.
# Current balance: 900
# Deposited 300.
# Current balance: 700
# Transferred 300 to Bob.
# Current balance: 900
# Current balance: 700
```

```python
# 04: IF/ELSE in a Comprehension

# Categorize numbers as "Even" or "Odd"
categories = []

for number in range(10):
    if number%2 == 0:
        categories.append("Even")
    else:
        categories.append("Odd")
        
print(categories)

# Using List Comprehension

categories_list_comprehension = [
   "Even" if number%2 ==0 else "Odd"  for number in range(10)
]
print(categories_list_comprehension)
```

### Analogy for Mutable Objects (Like Lists) and Immutable Objects (Like Tuples)

##### Mutable Objects (Like Lists)

Imagine you have a whiteboard (x) where you've written "1, 2." You invite your friend (y) to come and look at the whiteboard. Now, both of you are looking at the same board. If you erase the "1" and write "25" instead, your friend will immediately see the change because you're both looking at the same whiteboard.
##### Immutable Objects (Like Tuples)

Now, think of a printed poster (x) that says "1, 2." You make an identical copy of the poster and give it to your friend (y). If you decide to print a new poster that says "1, 2, 3," only your poster changes. Your friend's copy stays the same because each poster is a separate, unchangeable object.

##### Key Point

   **Mutable objects** (like lists): When you change it, anyone who has access to it sees the change because they all refer to the same object.
    
   **Immutable objects** (like tuples): When you "change" it, you're actually creating a new object. Others still see the original because their reference hasn’t changed.

    Lists (Mutable): If you change the content of x, y will reflect that change because both x and y point to

### reference vs memory address in python object

In Python, reference and memory address are related but distinct concepts:

    Reference: When you assign an object to a variable in Python, the variable holds a reference to the object, not the actual data. This means the variable is like a label that points to the object stored somewhere in memory.

    Memory Address: This is the actual location in the computer's memory where the object resides. Python manages memory addresses internally, so you don't typically deal with them directly.

Example:

    x = [1, 2, 3]
    y = x  # y references the same list object as x

    print(id(x))  # Returns the memory address of the list
    print(id(y))  # Same memory address as x

Both `x` and `y` reference the same object, and `id()` returns the memory address where this object is stored.

#### There is pointer in C or C++. Is there pointer in Python?

In Python, there aren't pointers like in C or C++. However, Python variables work similarly to pointers in that they reference objects in memory. When you assign a variable to another, you're copying the reference, not the actual data. This is why changes to mutable objects (like lists) through one variable are reflected in another that references the same object. However, Python abstracts away the low-level details of memory management, so you don't deal with pointers directly as you do in C or C++. 

### how to work with memory address in python?


In Python, you can work with memory addresses indirectly by using the `id()` function, which returns the memory address of an object. However, Python abstracts direct memory manipulation (unlike C/C++ where you can use pointers).

Example:

        x = [1, 2, 3]
        print(id(x))  # Outputs the memory address of the list object

        # Reassignment
        x = [123]
        print(id(x))  # Outputs the new memory address of the new list object

**Advanced:**

For more advanced use cases involving memory, you can use the `ctypes` module, which allows direct interaction with memory in a way similar to pointers in C. However, this is generally not common in everyday Python programming due to the language's high-level nature.

```python
x = [1, 2, 3]
y = x[0]  # y references the same list object as x[0]
print(y)
print(id(x))  # Returns the memory address of the list
print(id(y))  # memory address as x[0]

x[0]=5
y=x[0]
print(id(x))  # Returns the memory address of the list same to same the previous address
print(id(y))  # Returns the memory address of the new x[0] 
```

#### Python abstracts away the low-level details of memory management..what is abstracts here?

When we say Python "abstracts away the low-level details of memory management," it means that Python hides the complexity of managing memory (like allocating and freeing memory) from the programmer. You don't need to manually control memory allocation or deal with pointers as you would in languages like C or C++. Instead, Python automatically manages these details for you, allowing you to focus on writing your code without worrying about how memory is handled under the hood.
